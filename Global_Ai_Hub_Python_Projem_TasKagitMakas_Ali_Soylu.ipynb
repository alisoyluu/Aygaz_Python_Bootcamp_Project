```python
import time
import sys
import random

### Basit bir loading bar ile oyunu başlatmak için, yazılan fonksiyon; ###

def loading_animation(toplam_animasyon_zamani):
    toplam_bar_uzunlugu = 50  # Yükleme dolum çubuğu uzunluğu 50 karakter yapıldı.
    baslangic_zamani = time.time() # Animasyonun başlangıcı, baslangic_zamani değişkenine kaydedildi.

    while True:
        gecen_zaman = time.time() - baslangic_zamani # Animasyonun başlagıcından itibaren geçen süre hesaplanarak gecen_zaman değişkenine kaydedildi.
        if gecen_zaman > toplam_animasyon_zamani: # Geçen zaman, belirlenen animasyon zamanından büyükse; animasyon zamanını geçen zamana eşitleriz ki animasyon zamanını koruyabilelim. %120 gibi saçma değerler çıkamasın.
            gecen_zaman = toplam_animasyon_zamani

        # Dinamik bir şekilde karakterler ve boşlukları karıştırarak göstergede dolum efekti
        dolan_kismin_bar_uzunlugu = int(toplam_bar_uzunlugu * gecen_zaman / toplam_animasyon_zamani)
        bar = '#' * dolan_kismin_bar_uzunlugu + '-' * (toplam_bar_uzunlugu - dolan_kismin_bar_uzunlugu) #Fonksiyon, belirlenen süre boyunca bir yükleme çubuğunu dinamik olarak günceller. Süre boyunca # karakterleri yüklenme miktarını, - karakterleri ise henüz yüklenmemiş kısmı gösterir.
        yuzde_degeri = int(100 * gecen_zaman / toplam_animasyon_zamani)

        sys.stdout.write(f"\rYükleniyor: [{bar}] {yuzde_degeri}%")
        sys.stdout.flush() # sys.stdout.write() ile yazdırılan yükleme çubuğunun hemen eşzamanlı olarak ekranda görünmesini sağlamak için sys.stdout.flush() kullanıldı.

        if gecen_zaman >= toplam_animasyon_zamani:
            break

        time.sleep(0.8)  # Göstergeyi güncelleme süresi

    sys.stdout.write("\rYüklenme tamamlandı!                \n")

### Galibiyet, mağlubiyet ya da beraberlik durumuna göre metin ve ilgili emojisi ile sonucu belirtmek için yazılan fonksiyon; ###

def print_result_message(mac_sonucu):
    """Sonuç durumuna göre emoji gösterir."""
    if mac_sonucu == 'galibiyet':
        print("😊 Kazandınız!")
    elif mac_sonucu == 'maglubiyet':
        print("😢 Kaybettiniz!")
    elif mac_sonucu == 'beraberlik':
        print("⚖️ Berabere!")
    else:
        print("Sonuç belirsiz!")

### Oyunun tüm detaylarını barındıran fonksiyon ; ###

def tas_kagit_makas_Ali_SOYLU():
    print("Taş-Kağıt-Makas oyununa hoş geldiniz!")

    # Yükleme animasyonu göster
    loading_animation(5)  # 5 saniyelik bir yükleme animasyonu

    # Oyun kurallarını göster
    print("Kurallar:")
    print("Taş, Kağıt, Makas seçimlerinden birini yaparak oyuna başlayabilirsiniz.")
    print("Taş, Makas'ı keser. Bu durumda taş seçimini yapan 1 puan kazanır.")
    print("Kağıt, Taş'ı kaplar. Bu durumda kağıt seçimini yapan 1 puan kazanır.")
    print("Makas, Kağıt'ı keser.Bu durumda makas seçimini yapan 1 puan kazanır.")
    print("Her iki taraf da aynı seçimi yaparsa round beraberlik ile sonuçlanır. Kimseye puan yazılmaz.")
    print("İlk iki turu kazanan kişi oyunun galibi olacaktır.")
    print("Çıkmak için 'çıkış' yazabilirsiniz.")

    secenekler = ['taş', 'kağıt', 'makas']

    while True:
        kullanici_zaferi = 0
        bilgisayar_zaferi = 0

        while kullanici_zaferi < 2 and bilgisayar_zaferi < 2:
            kullanici_secimi = input("Taş, Kağıt veya Makas seçin: ").lower()
            if kullanici_secimi == 'çıkış':
                print("Oyundan çıkılıyor. Görüşmek üzere!")
                return
            if kullanici_secimi not in secenekler:
                print("Geçersiz seçenek. Lütfen 'Taş', 'Kağıt' veya 'Makas' seçin.")
                continue

            bilgisayar_secimi = random.choice(secenekler)
            print(f"Bilgisayarın seçimi: {bilgisayar_secimi}")

            if kullanici_secimi == bilgisayar_secimi:
                print_result_message('beraberlik')
            elif (kullanici_secimi == 'taş' and bilgisayar_secimi == 'makas') or \
                 (kullanici_secimi == 'kağıt' and bilgisayar_secimi == 'taş') or \
                 (kullanici_secimi == 'makas' and bilgisayar_secimi == 'kağıt'):
                print_result_message('galibiyet')
                kullanici_zaferi += 1
            else:
                print_result_message('maglubiyet')
                bilgisayar_zaferi += 1


            print(f"Skor - Siz: {kullanici_zaferi}, Bilgisayar: {bilgisayar_zaferi}")

        if kullanici_zaferi == 2:
            print("Tebrikler, oyunu kazandınız!")
            print_result_message('galibiyet')
        else:
            print("Üzgünüm, bilgisayar kazandı.")
            print_result_message('maglubiyet')

        tekrar_oyun_istegi = input("Başka bir oyun oynamak ister misiniz? (evet/hayır): ").lower()
        if tekrar_oyun_istegi == 'hayır':
            print("Oynadığınız için teşekkürler! Görüşmek üzere!")
            break
        elif tekrar_oyun_istegi != 'evet':
            print("Geçersiz giriş. Oyundan çıkılıyor.")
            break

# Oyunu çalıştırmak için aşağıdaki kodu kullanabilirsiniz
# tas_kagit_makas_Ali_SOYLU()
```
